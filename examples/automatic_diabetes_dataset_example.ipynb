```python
# Installing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# For this demo only — SaANN does not require scikit-learn.
from sklearn.datasets import load_diabetes 


# SaANN library for automatic
from saann.models import SequentialModel
```

```python
# Install scikit-learn in this notebook kernel if you need.
%pip install scikit-learn
```

```python
# Load dataset
diabetes = load_diabetes()
X = np.array(diabetes.data)
y = np.array(diabetes.target)
```

```python
# Initialize model
model = SequentialModel()
```

```python
# Define layer specifications: (input_size, neurons, activation, initialization)
layer_info = [
    (X.shape[1], 10, "relu", "he"),
    (10, 1, "linear", "he")
]
```

```python
# Run the automatic method
# Tip: keep real_time = False on notebooks
y_pred, final_pred_train, X_train, X_test, y_train, y_test = model.automatic(X=X, y=y, layers_info=layer_info, learning_rate=0.001, epochs=1000, batch_size=32, split_test_percentage=0.3, scaling='minmax', graphical=True, real_time=False, log_plot = True, test_loss = True, scatter_comparison = True)
```
